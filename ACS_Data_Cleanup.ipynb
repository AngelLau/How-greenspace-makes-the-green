```python
import pandas as pd
```

```python
df = pd.read_csv('ACS_16_5YR_S0601.csv')
df.head()
```

```python
df.columns = df.iloc[0,:]
df = df.iloc[1:,:]
df = df.reset_index(drop=True)
df.head()
```

## Clean Census tract

```python
def cleanCensusTract(x):
    """Cleans 'Geography' column into Census tract"""
    
    x_split = x.split(',')
    x_split = x_split[0].split(' ')
    return x_split[2]
```

```python
df['censusTract'] = df.Geography.apply(lambda x: cleanCensusTract(x))
```
